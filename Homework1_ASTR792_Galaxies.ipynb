#### b. Explain why the surface brightness of a galaxy does not depend on the galaxy's distance for nearby galaxies (where space-time) is locally flat

#### For nearby galaxies, the surface brightness of a galaxy is independent if spacetime is locally flat. To illustrate why this is the case, imagine one arcminute of a local galaxy. This patch of sky has a certain brightness due to the stars/objects that fit in that area. As the distance of that source increases, every single light emitting source gets fainter by the inverse square law $F \propto \frac{1}{d^2}$. But also, the physical scale spanned by each of the segment increases. The length of the segment is equal to $l=\theta d$. The physical area of that angular unit is proportional to $d^2$. If we put these together: $A = l^2=\theta^2d^2$ such that we can see $A \propto d^2$. Overall, this means that the effects cancel out and cause the surface brightness to be constant for local galaxies. Every piece of light emitting material is getting fainter by a factor of $\frac{1}{r^2}$ but at greater distances more sources are included in that angular unit (arcsecond). 

#### However, when we consider galaxies at high redshift, this relation no longer holds and surface brightness does depend on distance in that case due to cosmological effects such at photon arrival time, decreasing of photon energy due to wavelength change, scaling of theta and angular size.

#### a. If a galaxy has absolute magnitude M, derive an expression for the apparent magnitude m in terms of the redshift $z=V_r/c$, the absolute magnitude M, and a constant, C, which is the same for all objects.

#### Solution: 

#### Starting from the distance modulus equation we see a relation between the apparent and absolute magnitude.  

\begin{align}
    m_2 - m_1 &= 2.5log_{10}(\frac{L_1}{L_2}) \\
    m - M &= 2.5log_{10}(\frac{d}{10})^2 \\
\end{align}

#### We were able to subsittute in the fact that $m_2 = m$ apparent magnitude, $m_1 = M$ absolute magnitude, $d_2 = d$, and $d_1 = 10pc$ due to the definition of absolute magnitude. Additionally, we utilized the relation $L_1 = 4\pi d^2 F$ from the equation $F= \frac{L}{4\pi d}$.

#### Solving for apparent magnitude and using properties of logarithms we are able to rearrange the equation to the following: (Note: $log_{10}(10)=1$)
\begin{align}
    m &= M + 5log_{10} d -5log_{10}(10) \\
      &= M + 5log_{10} d -5
\end{align}

#### Now, if we utilize the fact that redshift $z=\frac{V_r}{c}$ we can make a substitution to the Hubble relation and substitute it into our equation for apparent magnitude. 

\begin{align}
    d &= h^{-1} \frac{V_r km*s^-1}{100} Mpc \\
      &= \frac{c}{100h} z km*s^{-1} Mpc
\end{align}

#### Substituting into our equation for apparent magnitude we get the following equation: 

\begin{align}
    m &= M + 5log_{10} d -5 \\
      &= M + 5log_{10} (\frac{c}{100h} z) - 5 \\
      &= M + 5log_{10} z + 5log_{10}(\frac{c}{100h}) - 5
\end{align}

#### So, finally, we get the relation:

\begin{align}
    m &= M + 5log_{10} z + C
\end{align}

#### Where $ C= 5log_{10}(\frac{c}{100h}) - 5$

# $Problem # 3$
#### The ALMA radio telescope can observe over a simultaneous frequency range of 8 GHz. Assume that you tune the frequency of the ALMA instrument (called a "correlator") to be centered to any observed frequency, e.g. that of a galaxy at rest in a cluster. The CO (2-1) spectral line has a rest-frame frequency of 230.538 GHz. You would liek to observed this line coming from galaxies in a cluster at z=1.60. How fast can a galaxy be moving along the line of sight w.r.t. the center of the cluster such that you will still be abel to detect the CO line? Express your answer in km/s.

#### Solution:
#### Knowing that $z=1.60$ and $\nu_{rest} = 230.538$ GHz, We can solve for the frequency at which we would observe the line emission on earth. In terms of redshift, one way to conceptualize this is to think of it as the factor by which our spectral lines are shifted. 
#### Equation (1) below relates emitted & observed frequency and redshift. Since we are given the rest frequency (or emitted frequency) and the value of the redshift, we can solve for the observed frequency as shown in the derivation in Eq. (2)

\begin{align}
     z+1 = \frac{\nu_e}{\nu_0} \label{eq:redshift} \tag{1} \\
\end{align}

\begin{align}
    \nu_0 &= \frac{\nu_e}{z+1} \tag{2} \\
          &= \frac{230.538 GHz}{1.60+1} \\
          &= \boxed{88.66 GHz}
\end{align}

#### Now, in order to solve for the velocity of the galaxy along the line of sight we can use the following relation to solve for the velocity. Equation (3) shows how the relativistic velocity, or true line-of-sight velocity, which is related to the observed and rest frequancies. I utilized the NRAO Documentation for VLA/ALMA for advice on solving this. https://science.nrao.edu/facilities/vla/docs/manuals/obsguide/modes/line There is an approximation that can be used instead of the full equation (3) where $V^{optical}=\frac{\lambda - \lambda_0}{\lambda_0}c = cz$ where z is the redshift of the source;  and 0 are the corresponding observed and rest wavelengths, respectively. However, I opted to use the more "cumbersome" equation in order to solve for the line of sight velocity. 

\begin{align}
    V = \frac{\nu_0^2 -\nu^2}{\nu_0^2 +\nu^2}c \tag{3}
\end{align}

#### where $\nu$ is the observed frequency, $\nu_0$ is the rest frequency of the spectral line (in this case CO (2-1)), c is the speed of light, and V is the line-of-sight velocity of the source. This equation is derived from the relativistic doppler shift equation given below for completeness though in terms of wavelength ($\lambda = \frac{c}{\nu}$). 

\begin{align}
    {\lambda }_{obs} ={\lambda }_{s}\sqrt[]{\frac{1+\frac{v}{c}}{1-\frac{v}{c}}} \tag{4}
\end{align}

#### Solving for the line of sight velocity: 

\begin{align}
    V &= \frac{\nu_0^2 -\nu^2}{\nu_0^2 +\nu^2}c \tag{5} \\
      &= \frac{230.538^2 - 88.66^2}{230.538^2 + 88.66^2}c \\
      &= 0.74c \\
      &= \boxed{2.2 \times 10^5 {km/s}}
\end{align}

#### Thus this is the speed of the galalxy moving a long the line of sight (w.r.t the center of the cluster) such that you can still detect the CO (2-1) spectral line.

#### Compute the comoving volume in the two intervals $0.05<z<1$ and $1<z<2$ tha tis contained within a field whose solid angle is $0.25 deg^2$. Assume $\Omega_m = 0.3$, $\Omega_\Lambda = 0.7$, and $h=0.7$. I want you to write a program to solve this problem by numerically integrating E(z).

You must Include your code with this assignment, which you can upload as a Jupyter Notebook or Python code.

\begin{align}
    E(z) = \sqrt{\Omega_M(1+z)^3 + \Omega_k(1+z)^2 + \Omega_\Lambda} \tag{6}
\end{align}

\begin{align}
    dV_c = D_H \frac{(1+z)^2 D_A^2}{E(z)}d\Omega dz \tag{7}
\end{align}

#### where $D_A$ is the angular diameter distance at redshift z, $D_H$ is the Hubble distance 

\begin{align}
    dV_c &= D_H \frac{(1+z)^2 D_A^2}{E(z)} d\Omega dz \\
         &= D_H \frac{D_c^2}{E(z)} d\Omega dz \\
         &= D_H^3 \frac{[\int^z_0 \frac{1}{E(z')}dz']^2}{E(z)} d\Omega dz \\
         &= \frac{c^3}{H_o^3} \frac{[\int^z_0 \frac{1}{E(z')}dz']^2}{E(z)} d\Omega dz \\
\end{align}

#### and to take the integral we can use numerical integration -- I chose numpy.trapz  to do trapezoidal -- since it does not depend on $\Omega$ I chose to multiply that at the very end. Below you will find the code I wrote in order to do this. 

```python
import numpy as np
from numpy import trapz
import math
```

```python
# Define E(z) fucntion 
def get_E(z):
    E = math.sqrt((0.3 * (1 + z)**3) + 0.7)
    return E
```

```python
# Define H_inv
def get_H_inv(z):
    
    # Make array to append to
    E_list = []
    tot_z_range = []
    Ho = 100 * 0.7 # km s^-1 Mpc^-1
    
    for z in np.arange(0, z, 0.0001):
        val = 1 / (get_E(z) *  Ho)
        E_list.append(val)
        tot_z_range.append(z)
        
    k = np.trapz(E_list, tot_z_range) # Perform numerical integration and return the value
    
    return k     
```

```python
# Define function
def int_vol(z_initial,z_final):
    
    # Make an empty array to append to 
    dV_arr = []
    z_range = []
    
    # Constants 
    c = 3 * (10**5) # km/s
    Ho = 100 * 0.7 # km s^-1 Mpc^-1
    
    for z in np.arange(z_initial,z_final, 0.0001):
        G = ((c**3) / Ho) * (1 / (get_E(z))) * (get_H_inv(z) ** 2)
        dV_arr.append(G)
        z_range.append(z)
        
    v = np.trapz(dV_arr, z_range)
    
    ster = 7.6154 * (10 ** (-5)) # Convert deg^2 to steradians
    v_final = v * ster   
    
    return v_final
```

```python
# Intervals to compute comoving volume over
z_initial = 0.05
z_final = 1

#z_initial = 1
#z_final = 2
```

```python
x = int_vol(z_initial,z_final)
print(x)
```
